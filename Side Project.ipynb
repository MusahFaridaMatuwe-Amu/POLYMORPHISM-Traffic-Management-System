## Case Study Activity: Smart Traffic Management System

A smart city uses intelligent traffic devices. Every device receives the command `activate()`, but each performs a different task.

### Requirements:
- Create:
  - **Parent Class**: `TrafficDevice`
  - **Child Classes**: `TrafficLight`, `SpeedCamera`, `PedestrianSignal`
- Each class must override: `activate()`
- Create one object of each class, store them in a list, and activate them without checking their types.
- Add an `EmergencySiren` class without modifying the activation loop.

```python
class TrafficDevice:
    """Parent class for all smart traffic devices."""

    def activate(self):
        print("Generic traffic device activated.")


class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light activated: Cycling through Red -> Green -> Yellow.")


class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera activated: Monitoring vehicle speed.")


class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal activated: Displaying Walk/Don't Walk signal.")


# New class added WITHOUT modifying the activation loop below
class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren activated: Alerting nearby vehicles to give way.")


# Create one object of each class and store them in a list
devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal(),
    EmergencySiren()
]

# Activate every device polymorphically (no type checking)
for device in devices:
    device.activate()
```
